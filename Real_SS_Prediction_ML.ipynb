## Membaca Dataset

```python
def open_dataset (file_dataset):
  raw_primer = []
  raw_sekunder = []

  with open (file_dataset, 'r') as f:
    for count, line in enumerate(f, start =1):
      if count % 2 == 0 :
        raw_sekunder.append(line.strip())
      else:
        raw_primer.append(line.strip())
  return raw_primer, raw_sekunder
```

## Split Sekuens

```python
def split (raw):
  return [char for char in raw]

def split_raw (raw_primer, raw_sekunder):
  split_primer = []
  split_sekunder = []
  for i in range(len(raw_primer)):
    split_primer.append(split(raw_primer[i]))
    split_sekunder.append(split(raw_sekunder[i]))
  return split_primer, split_sekunder
```

## Padding dan Sliding Window

```python
def window_padding_data(sequence, size, size_feature):
  num = int(size/2)
  zeros = [0]* size_feature
  for i in range(len(sequence)):
    # print (i)
    for j in range(num):
      sequence[i].append(zeros)
      sequence[i].insert(0, zeros)

  X = []
  temp = []

  for k in range(len(sequence)):
    for l in range(len(sequence[k])-(size - 1)):
      temp = sequence[k][l:l+size]
      X.append(temp)
      temp = []
  return X
```

## Ektrkasi Fitur

```python
def Ekstrak_Fitur_VHSE(split_primer, size, size_feature):
  kimia_fisik = VHSE(split_primer)
  window = np.array(window_padding_data(kimia_fisik, size, size_feature))
  windows = window.reshape(sum(len(sublist) for sublist in split_primer), size * size_feature) # 8 terdiri dari fitur VHSE
  return windows
```

```python
def Ekstrak_Fitur_VHSE_PC(split_primer, size, size_feature):
  kimia_fisik = VHSE_PC(split_primer)
  window = np.array(window_padding_data(kimia_fisik, size, size_feature))
  windows = window.reshape(sum(len(sublist) for sublist in split_primer), size * size_feature) # 8 terdiri dari fitur VHSE
  return windows
```

## Encode Struktur Sekunder Protein

```python
import numpy as np


def encode_sekunder(split_sekunder):
  encoded_sekunder = []
  switch2 = {
      'H' : 0,
      'G' : 1,
      'I' : 2,
      'E' : 3,
      'B' : 4,
      'S' : 5,
      'T' : 6,
      'C' : 7,
      'P' : 8
  }
  distinct_of_string = ''
  for i in range(len(split_sekunder)):
    seq2 = split_sekunder[i]
    y = []
    for j in range(len(seq2)):
      # print(j)
      x = switch2[seq2[j]]
      y.append(x)

    encoded_sekunder.append(y)

  return encoded_sekunder

def target(encoded_sekunder):
  Y = []
  for i in range(len(encoded_sekunder)):
    for j in range(len(encoded_sekunder[i])):
      Y.append(encoded_sekunder[i][j])
  return Y
```

## Decode Struktur sekunder

```python
def decode_sekunder(prediction):
  dict = {
      0 : 'H',
      1 : 'G',
      2 : 'I',
      3 : 'E',
      4 : 'B',
      5 : 'S',
      6 : 'T',
      7 : 'C',
      8 : 'P'
  }
  hasil_decode = ''.join([dict[i] for i in prediction])
  return hasil_decode
```

## Evaluasi Model

```python
def evalute (y_test, y_pred):
  score = 0
  for i in range(len(y_test)):
    if y_test[i] == y_pred[i]:
      score += 2
    else:
      score -= 1
  return score

def kemiripan (evaluate, y_test):
  akurasi = (evaluate / (2*len(y_test))) * 100
  return akurasi
```
